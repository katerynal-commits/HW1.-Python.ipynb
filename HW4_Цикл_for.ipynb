**1. Цикл for - базовий**

Використовуючи цикл `for`, виведіть числа від 1 до 7 включно використовуючи функцію range.


```python
#цикл по діапазону

for i in range (1,8):
  print (i)
```

**2. Цикл for по списку**

Дано список:

```python
names = ["Олег", "Анна", "Іван", "Марія"]
```



---


Використовуючи цикл, виведіть кожне ім'я з привітанням: "Привіт, [ім'я]!"


```python
names = ["Олег", "Анна", "Іван", "Марія"]
for name in names:
  print (f"Привіт,{name}!")
```

**3. Цикл з enumerate**

Дано список оцінок:

```python
grades = [85, 92, 78, 95, 88]
```


Використовуючи `enumerate`, виведіть кожну оцінку з номером: "Студент 1: 85 балів"


```python
grades = [85, 92, 78, 95, 88]
for i, grade in enumerate (grades, start=1):
  print(f"Cтудент {i}: {grade} балів")
```

**4. Сума парних чисел**

Напишіть програму, яка обчислює суму всіх парних чисел від 1 до 20 включно.

```python
sum_even = 0
for i in range(1, 21):
    if i % 2 == 0:
        sum_even += i

print("сумма всіх парних чисел від 1 до 20 включно:",sum_even)
```
